Funcion crear_un_grafo_completo: Está función recibirá un entero (n) y devolverá un diccionario que representa la gráfica Kn.

```python
n = int(input('Tamaño de la grafica: '))  #solicitamos el entero para crear la grafica

lista_nod = list()  #creamos una lista vacia
for i in range(1,n+1): #Esto genera la secuencia basada en el valor de inicio y parada
    lista_nod.append(i) #Agrega un elemento a la lista
    
```

```python
grafica = {}  #creamos un diccionario vacia

for i in range(1,n+1): #Esto genera la secuencia basada en el valor de inicio y parada
    lista_tprl =list(lista_nod)  #creamos una lista temporal la cual llenaremos con los valores guardados ateriormente
    lista_tprl.remove(i) #elimina el primer elemento coincidente (que se pasa como argumento) de la lista.
    llave= {i:lista_tprl}  #ira colocando el nodo como llave y el tamaño del valor como el nuevo valor.
    grafica.update(llave) #Inserta un elemento en el diccionario.

print(grafica)  #Imprimimos la grafica
```

Funcion calcula_grados_de_entrada:Esta función recibirá un diccionario (d_grafica) representando una gráfica dirigida y devolverá un diccionario en donde las llave representan los nodos de (d_grafica), y los valores el grado de entrada de cada
nodo

```python
d_grafica = {1:[2], 2:[4], 3:[2], 4:[1,3,5], 5:[2,4,6], 6:[1]} #creamos diccionario de la fig. 5

grado_nod= {} #creamos diccionario vacio

for i in d_grafica: 
    llave = {i:len(d_grafica[i])} # cuenta el grado de entrada por nodo en la lista
    grado_nod.update(llave)  #Inserta un elemento en el diccionario vacio

print(grado_nod) #Imprimimos la diccionario
```

Funcion dist_grados_de_entrada. Esta función recibirá un diccionario (d_grafica) representando una gráfica dirigida y devolverá un diccionario (distr) en donde las llaves representan los diferentes grados de entrada, y los valores representan el número de nodos con dicho grado. Esta función debe utilizar la función calcula_grados_de_entrada.

```python
distr = {}  #creamos diccionario distr
lista_tprl2= list() #creamos una lista temporal

for A in grado_nod.values():  #devuelve un objeto de vista que muestra una lista de todos los valores en el diccionario.
    lista_tprl2.append(A) #Agrega un elemento a la lista 
    lista_tprl2.sort() #Ordena los valores de la lista 
    
for i in lista_tprl2:
    llaves = lista_tprl2.count(i) #Devuelve el número de veces que aparece el valor en la lista
    llaves2 = {llaves:i} #ira colocando llave tomara el valor del nodo
    distr.update(llaves2)  #Inserta un elemento en el diccionario.
print(distr)
```

Funcion dist_normalizada. Esta función recibirá un diccionario (d_grafica) representando una gráfica dirigida y devolverá un diccionario (distr_norm) en donde las llaves representan los diferentes grados de salida, y los valores representan el número
de nodos con dicho grado. Esta función debe llamar a la función dist_grados_de_entrada.

```python
distr_norm = {} #creamos diccionario distr_norm

for i in distr:
    llave = distr[i] # llave toma el valor de la lista
    valores= i # valores toma el valor de i
    llaves3 = {llave:valores} #llaves3 tomara los valores llave y valores
    distr_norm.update(llaves3)#Inserta un elemento en el diccionario.
print(distr_norm)
```

Funcion grafica_distr. Esta función recibirá un diccionario (distr) representando la distribución de grados de una gráfica, y graficará dicha distribución usando la librería matplotlib. El eje x debe estar etiquetado como ‘grado nodal’, el eje y como
‘número de nodos’, el título debe ser ‘Distribución de Grados de Entrada’.


```python
import matplotlib.pyplot as plt

x= list() #creamos lista vacia
y= list() #creamos lista vacia

for i in distr:
    x.append(distr[i]) #agregamos elementos del diccionario distr a lista 
    y.append(i) #agregamos elementos a lista 
plt.plot(x,y)
plt.title('Distribución de Grados de Entrada') #titulo de la grafica
plt.xlabel('Grado nodal') #etiqueta de eje x
plt.ylabel('Num de nodos') #etiqueta de eje y
plt.show() #se usa para mostrar todas las figuras.
```
